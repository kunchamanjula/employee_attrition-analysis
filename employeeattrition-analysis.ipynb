# Job Involvement vs Attrition Rate:

```python
attrit = df.groupby(['JobInvolvement','Attrition'])["EmployeeID"].sum().unstack()
attrit
```

```python

fig, ax = plt.subplots()

colors = ['#24b1d1', '#ae24d1']
bottom = len(attrit)

for i, col in enumerate(attrit.columns):
  ax.bar(attrit.index, attrit[col], bottom=bottom, label=col,
         color=colors[i])
  bottom += np.array(attrit[col])
ax.set_title('job involvement vs attrition')
ax.legend()
plt.show()
```

**Observation:**
* By this plot we can see that mostly the attrition rate decraesing with the increase in job invovlement.
* here we have 4 levels of job involvement out of these level1 and level4 are low in attrition rate.
* but when we observe in level3 they are most involved in jobs but their attrition rate is more.

# Department wise Attrition Rate vs work life balance

```python
attcount = df.groupby(["Department","Attrition"])["WorkLifeBalance"].sum().unstack()
attcount
```

```python
fig, ax = plt.subplots()
# First plot the "Department" bars for every day.
ax.bar(attcount.index, attcount["No"], label='no')
# Then plot the 'Attrition' bars on top, starting at the top of the 'Department'
# bars.
ax.bar(attcount.index, attcount["Yes"], bottom=attcount["No"],
       label='yes')
ax.set_title('Department wise attrition rate vs worklife balance ')
ax.legend()
plt.show()
```

*Obsservation:*
*  By the graph we can see that the employees who works in research & development department  has better work life balance among these.But their attritaion rate is bit higher.
* But when we statstically calcualtes the Human resources department employees have high attrition rate .
* by this we can say that the work life balances play major role in attrition rate.

# Business travel wise department vs attrition 

```python
df["Department"].value_counts()
```

```python
def department(i):
    for i in df["Department"]:
        if i == "Research & Development":
            return 1
        elif i=="Sales":
            return 2
        else:
            return 3
```

```python
df["Department"]=df["Department"].apply(department)
```

```python
travel = df.groupby(["BusinessTravel","Attrition"])["Department"].sum().unstack()
travel
```

```python

fig, ax = plt.subplots()

colors = ['#a83273', '#32a868']
bottom = len(travel)

for i, col in enumerate(travel.columns):
  ax.bar(travel.index,travel[col], bottom=bottom, label=col,
         color=colors[i])
  bottom += np.array(travel[col])
ax.set_title('BusinessTravel vs attrition')
ax.legend()
```

**observation:**
* by the above plot we can see that the employee who travels rarely wants to attritite from company.
* the non travel employee is least likely to attritiate from the company.
